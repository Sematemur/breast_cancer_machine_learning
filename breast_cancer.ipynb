```python
import pandas as pd   #breast cancer datasetimi yükledim.
import numpy as np
df=pd.read_csv('/content/breast-cancer.csv')
df
```

```python
df['diagnosis'].unique     #burda bağımlı değişkenimin sayısını ve her birinden kaç tane var ona baktım.
df['diagnosis'].value_counts()
```

```python
pd.get_dummies(df,columns=['diagnosis'],drop_first=True).head()   #one hot encoding yaptım ve M B değerlerini 1 ve 0 yaptım
```

```python
df.columns
```

```python
X=df.drop('diagnosis',axis=1)    #bağımlı ve bağımsız değişkenlerimi belirledim
Y=df['diagnosis']
df
```

```python
from sklearn.linear_model import LogisticRegression                  #logisticregression kurdum ve train test oluşturdum
model=LogisticRegression()
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
model.fit(X_train,y_train)
```

```python
predicts=model.predict(X_test)   #tahmin ettim
```

```python
from sklearn.metrics import confusion_matrix    #confusion matrixini cıkardım
matrıx=confusion_matrix(y_test,predicts)
matrıx
```

```python
import matplotlib.pyplot as plt     #confusion matrix'i görselleştirdim
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.heatmap(matrıx, annot=True, fmt='d', cmap='PuRd',
            xticklabels=['1', '0'],
            yticklabels=['1', '0'])
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import classification_report    #burda da accuracy değerime baktım. tahminimin ne kadar doğru olduğu ile ilgili.
print(classification_report(y_test,predicts))
```
